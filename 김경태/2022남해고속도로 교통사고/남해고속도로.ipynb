```python
import os
import pandas as pd
import matplotlib.pyplot as plt

#     
folder_path = '2022 csv '

#   
accident_counts = {}

#     
files = os.listdir(folder_path)

#    
for file in files:
    if file.endswith('.csv'):
        #  
        file_path = os.path.join(folder_path, file)
        
        #  
        df = pd.read_csv(file_path, encoding='cp949')
        
        #  
        highway_name = file.split('2022')[1].split('.')[0]
        
        #   
        accident_count = len(df)
        
        #    
        if highway_name in accident_counts:
            accident_counts[highway_name] += accident_count
        else:
            accident_counts[highway_name] = accident_count

#   
data = pd.DataFrame(list(accident_counts.items()), columns=['', ''])

#      ()
data = data.sort_values(by='', ascending=True)  # ascending True   

#   (x y )
plt.figure(figsize=(12, 8))
plt.barh(data[''], data[''])  # barh     
plt.xlabel(' ')
plt.ylabel('')
plt.title('2022   ')
#      
for index, value in enumerate(data['']):
    plt.text(value, index, str(value), ha='left', va='center', fontsize=10)
plt.tight_layout()
plt.show()

#  10  
top_10_data = data.tail(10)  # tail   10 

#    10    
plt.figure(figsize=(12, 8))
plt.pie(top_10_data[''], labels=top_10_data[''], autopct='%1.1f%%', startangle=140)
plt.title('2022    TOP 10')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#   
plt.rcParams['font.family'] = 'Malgun Gothic'

#  
file_path = './.csv'
df = pd.read_csv(file_path, delimiter=',', encoding='cp949')

# ''  datetime  
df[''] = pd.to_datetime(df[''], format='%Y %m %d %H')


#     
dfs_by_gu = {gu: df[df[''].str.contains(gu)] for gu in cities_ending_with_si}

#  
fig, ax = plt.subplots(figsize=(12, 8))

heatmap_data = pd.DataFrame({gu: df_gu.groupby(df_gu[''].dt.hour).size() for gu, df_gu in dfs_by_gu.items()})

sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5)
plt.title('       ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
