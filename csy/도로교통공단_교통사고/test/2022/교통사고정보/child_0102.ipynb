```python
#    
import matplotlib.pyplot as plt
import pandas as pd

age_groups = ["20", "21-30", "31-40", "41-50", "51-60", "61-64", "65", ""]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

#  
mortality_rate = [death / accident * 100 if accident > 0 else 0 for death, accident in zip(death_counts, accident_counts)]

df = pd.DataFrame({
    '': age_groups,
    '': mortality_rate
})

#  
plt.figure(figsize=(10, 6))
plt.scatter(df[''], df[''], s=df['']*10, alpha=0.5, color='coral')
plt.title('   ')
plt.xlabel('')
plt.ylabel(' (%)')
plt.show()

#  
plt.figure(figsize=(8, 8))
plt.pie(df[''], labels=df[''], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('   ')
plt.show()

#  
plt.figure(figsize=(10, 6))
plt.bar(df[''], df[''], color='skyblue')
plt.title('   ')
plt.xlabel('')
plt.ylabel(' (%)')
plt.show()

#  
plt.figure(figsize=(10, 6))
plt.plot(df[''], df[''], marker='o', color='orange')
plt.title('   ')
plt.xlabel('')
plt.ylabel(' (%)')
plt.show()

```

```python
#  -  

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

age_groups = ["20", "21-30", "31-40", "41-50", "51-60", "61-64", "65", ""]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

#   
df = pd.DataFrame({
    '': age_groups,
    '': accident_counts,
    '': death_counts,
    '': injury_counts
})

plt.figure(figsize=(10, 6))
sns.heatmap(df.set_index('').T, annot=True, cmap='YlGnBu', fmt='d', linewidths=.5, cbar=True)
plt.title('   ')
plt.show()
```
