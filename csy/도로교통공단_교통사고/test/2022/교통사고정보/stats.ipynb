```python
#   - 3   
#   =  ,   = ,   = 

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc 
rc('font', family='AppleGothic') 		
plt.rcParams['axes.unicode_minus'] = False  


age_groups = ["20", "21-30", "31-40", "41-50", "51-60", "61-64", "65", ""]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

#  
plt.figure(figsize=(10, 6))
plt.scatter(age_groups, accident_counts, s=np.array(injury_counts) / 100, c='blue', alpha=0.5, label=' ')
plt.scatter(age_groups, death_counts, s=np.array(death_counts) * 2, c='red', alpha=0.5, label=' ')

plt.title('    ( )')
plt.xlabel('')
plt.ylabel(' ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#  
age_groups = ["20", "21-30", "31-40", "41-50", "51-60", "61-64", "65", ""]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

#   
df = pd.DataFrame({
    '': age_groups * 3,
    '': [''] * 8 + [''] * 8 + [''] * 8,
    '': accident_counts + death_counts + injury_counts
})

#  
plt.figure(figsize=(12, 6))
sns.boxplot(x='', y='', hue='', data=df)
plt.title('    -  ')
plt.xlabel('')
plt.ylabel('')
plt.show()


#  
plt.figure(figsize=(12, 6))
sns.lineplot(x='', y='', hue='', data=df)
plt.title('    -  ')
plt.xlabel('')
plt.ylabel('')
plt.show()

#  
plt.figure(figsize=(12, 6))
sns.barplot(x='', y='', hue='', data=df)
plt.title('    -  ')
plt.xlabel('')
plt.ylabel('')
plt.show()

```
