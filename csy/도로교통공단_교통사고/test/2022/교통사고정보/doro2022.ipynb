```python
#    
import matplotlib.pyplot as plt

age_groups = ["20", "21-30", "31-40", "41-50", "51-60", "61-64", "65", ""]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

#  
plt.figure(figsize=(15, 7))
plt.subplot(1, 2, 1)
plt.pie(accident_counts, labels=age_groups, autopct='%1.1f%%', startangle=90)
plt.title('    ')

#  
plt.subplot(1, 2, 2)
bar_width = 0.3
index = range(len(age_groups))
plt.bar(index, death_counts, width=bar_width, label='')
plt.bar([i + bar_width for i in index], injury_counts, width=bar_width, label='')
plt.xlabel('')
plt.ylabel('')
plt.title('      ')
plt.xticks([i + bar_width/2 for i in index], age_groups)
plt.legend()

#  
plt.tight_layout()
plt.show()
```
