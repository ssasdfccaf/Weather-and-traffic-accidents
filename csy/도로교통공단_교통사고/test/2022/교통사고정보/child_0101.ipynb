```python
import matplotlib.pyplot as plt
from matplotlib import rc 
rc('font', family='AppleGothic') 		
plt.rcParams['axes.unicode_minus'] = False  


data = {
    "": [" ", " ", " ", "", ""],
    "": [3, 0, 1, 14, 0],
    "": [7367, 38, 1106, 2838, 40]
}

# pie  ->      
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6)) #  

#   -   
axes[0].pie(data[""], labels=data[""], autopct='%1.1f%%', startangle=90)
axes[0].set_title('')

#   -   
axes[1].pie(data[""], labels=data[""], autopct='%1.1f%%', startangle=90)
axes[1].set_title('')

plt.show()
```
