```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rc
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False  


# CSV 파일을 읽어옴
data = pd.read_csv('../서울예측DB2(수정).csv', encoding='cp949')

# 그래프를 그리기 위한 설정
plt.figure(figsize=(12, 8))

# 어린이 사고 수를 막대 그래프로 표시
sns.barplot(x='지역구', y='어린이사고수', data=data, palette='viridis')
plt.title('서울시 지역구별 어린이 사고 수')
plt.xlabel('지역구')
plt.ylabel('어린이 사고 수')

# 그래프 표시
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
