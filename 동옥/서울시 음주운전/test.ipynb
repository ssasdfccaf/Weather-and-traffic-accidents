```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

#   
font_path = "C:/Windows/Fonts/malgun.ttf"  #    
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV  
file_path_accident = ' 21,22 .csv'
file_path_traffic = ' 21,22 .csv'

df_accident = pd.read_csv(file_path_accident, encoding='cp949')
df_traffic = pd.read_csv(file_path_traffic, encoding='cp949')

# X Y 
x_column = ''
y_column_accident_2021 = '2021'
y_column_accident_2022 = '2022'
y_column_traffic_2021 = '2021'
y_column_traffic_2022 = '2022'

#   
df_accident[''] = ((df_accident[y_column_accident_2022] - df_accident[y_column_accident_2021]) / df_accident[y_column_accident_2021]) * 100

#   
df_traffic[''] = ((df_traffic[y_column_traffic_2022] - df_traffic[y_column_traffic_2021]) / df_traffic[y_column_traffic_2021]) * 100

#     
df_difference = pd.DataFrame({
    '': df_accident[x_column],
    '': abs(df_accident[''] - df_traffic[''])
})

#    
df_difference = df_difference.sort_values(by='', ascending=False)

#  
plt.figure(figsize=(15, 8))

#     
bar_colors = ['lightblue' if diff < 0 else 'lightcoral' for diff in df_difference['']]
plt.bar(df_difference[''], df_difference[''], color=bar_colors, label='')
plt.title('       (2021-2022)')
plt.xlabel('')
plt.ylabel('   (%)')
plt.xticks(rotation=45)

#  
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV  
file_path_cctv = ' cctv.csv'
file_path_traffic = ' 21,22 .csv'

df_cctv = pd.read_csv(file_path_cctv, encoding='cp949')
df_traffic = pd.read_csv(file_path_traffic, encoding='cp949')

#   ''  
df_merged = pd.merge(df_cctv[['', 'cctv']], df_traffic[['', '2022']], left_on='', right_on='', how='inner')

#  
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cctv', y='2022', data=df_merged, hue='', palette='viridis', s=100)

plt.title(' CCTV  2022   ')
plt.xlabel(' CCTV ')
plt.ylabel('2022  ')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
