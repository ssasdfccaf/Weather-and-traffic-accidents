```python
import matplotlib.pyplot as plt

#  
data = {
    "": ["", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", ""],
    "": [79.09, 78.43, 76.51, 79.07, 79.28, 78.16, 82.64, 76.65, 68.12, 79.79, 64.24, 67.96, 74.17, 74.16, 74.09, 75.24, 69.08],
    "": [35, 42, 57, 36, 34, 46, 7, 6, 24, 3, 28, 25, 12, 47, 13, 9, 23]
}

#  
fig, ax1 = plt.subplots(figsize=(15, 7))

#   
ax1.plot(data[""], data[""], marker='o', color='skyblue', label='')
ax1.set_xlabel('')
ax1.set_ylabel('', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('  ')

#    (  )
ax2 = ax1.twinx()
ax2.plot(data[""], data[""], marker='o', color='salmon', label='')
ax2.set_ylabel('', color='salmon')
ax2.tick_params(axis='y', labelcolor='salmon')

#  
plt.show()
```
